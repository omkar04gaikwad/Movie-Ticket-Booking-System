```python
import random
from datetime import datetime, timedelta
import pymysql
```

```python
movies = movie_data[["id", "original_title"]].to_dict(orient="records")
theaters = [
    "AMC Boston Common", "Regal Fenway", "Cinemark NYC", 
    "ArcLight Chicago", "Alamo Drafthouse LA"
]
cities = ["Boston", "NYC", "Chicago", "Los Angeles", "San Francisco"]
show_times = ["10:00 AM", "1:30 PM", "5:00 PM", "8:30 PM"]
showtime_data = []
seat_data = []
```

```python
from datetime import datetime, timedelta

# Convert 12-hour format to 24-hour format
def convert_to_24_hour(time_str):
    return datetime.strptime(time_str, "%I:%M %p").strftime("%H:%M:%S")  # Converts to 24-hour format
```

```python
for movie in movies:
    for _ in range(3):  # Generate 3 random showtimes per movie
        theater = random.choice(theaters)
        city = random.choice(cities)
        date = datetime.today() + timedelta(days=random.randint(0, 7))  # Next 7 days
        time = convert_to_24_hour(random.choice(show_times))  # Convert to 24-hour format
        showtime_data.append({
            "movie_id": movie["id"],
            "movie_title": movie["original_title"],
            "theater_name": theater,
            "city": city,
            "show_date": date.strftime("%Y-%m-%d"),
            "show_time": time
        })

        # Generate Seat Availability for Each Showtime
        showtime_id = len(showtime_data)  # Temporary showtime_id
        for seat in range(1, 51):  # 50 seats per showtime
            seat_data.append({
                "showtime_id": showtime_id,
                "seat_number": f"{chr(65 + (seat // 10))}{seat % 10}",  # A1, B2, etc.
                "status": random.choice(["Available", "Booked"])
            })

# Step 4: Convert to DataFrames
showtime_df = pd.DataFrame(showtime_data)
seat_df = pd.DataFrame(seat_data)
```

```python
mysqlconnection = pymysql.connect(
    host="localhost",
    user="root",
    password="hmyHISCGzUULfLLg",
    database="MovieBooking"
)
cursor = mysqlconnection.cursor()
```

```python
# Insert showtime data
for _, row in showtime_df.iterrows():
    cursor.execute("""
        INSERT INTO showtimes (movie_id, movie_title, theater_name, city, show_date, show_time)
        VALUES (%s, %s, %s, %s, %s, %s)
    """, (row["movie_id"], row["movie_title"], row["theater_name"], row["city"], row["show_date"], row["show_time"]))

# Insert seat data
for _, row in seat_df.iterrows():
    cursor.execute("""
        INSERT INTO seats (showtime_id, seat_number, status)
        VALUES (%s, %s, %s)
    """, (row["showtime_id"], row["seat_number"], row["status"]))
```
